Data Description

For this project, I will use the Foursquare API to compare the data of NYC and Toronto by neighborhood. The data will include information about each neighborhood like museums, restaurants, hotels, significant landmarks, galleries, and other tourist attractions. For NYC, I will investigate the borough of Manhattan. For Toronto, I will investigate Downtown Toronto. I will use clustering to segment the neighborhoods with similar objects. This will help locate tourist hubs and will allow me to judge the similarity or dissimilarity of the two cities.
